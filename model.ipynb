```python
small_data = data.head(100);
```

```python
def get_vector_df(col) :
    desc_vec = []
    for text in small_data[col]:
        filtered_text = ""
        doc = nlp(text)
        for token in doc:
            if(token.is_stop==False):
                filtered_text = filtered_text+" "+str(token)
        doc = nlp(filtered_text)
        desc_vec.append([doc.vector])
    df = pd.DataFrame(desc_vec)
    return df
```
